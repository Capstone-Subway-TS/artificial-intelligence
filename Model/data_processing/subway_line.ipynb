```python
import heapq

def dijkstra(graph, start, end):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    previous_nodes = {node: None for node in graph}
    queue = []
    heapq.heappush(queue, [distances[start], start])

    while queue:
        current_distance, current_node = heapq.heappop(queue)

        if current_node == end:
            break

        if distances[current_node] < current_distance:
            continue

        for adjacent, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[adjacent]:
                distances[adjacent] = distance
                previous_nodes[adjacent] = current_node
                heapq.heappush(queue, [distance, adjacent])

    path = []
    step = end
    while step is not None:
        path.append(step)
        step = previous_nodes[step]
    path.reverse()

    return path
```

```python
import heapq

def calculate_wait_time(current_minute, frequency=10):
    return (frequency - current_minute % frequency) % frequency

# 현재 시간 예시 (분으로 계산)
current_time = 8 * 60  # 8시간을 분으로 변환

# 기본 환승 시간 설정
transfer_penalty = 3  # 환승에 추가되는 시간은 3분

# 그래프 생성
subway_graph = {}

for line in lines:
    for station in line:
        if station['name'] not in subway_graph:
            subway_graph[station['name']] = {}
        # 다음 역 연결
        if station['next'] != "None":
            next_station = station['next'] if type(station['next']) is not list else station['next'][0]
            subway_graph[station['name']][next_station] = 2  # 가정된 이동 시간
            if next_station not in subway_graph:
                subway_graph[next_station] = {}
            subway_graph[next_station][station['name']] = 2

        # 환승역 처리
        if station['transfer'] == "True":
            for line_code in station['transline'] if isinstance(station['transline'], list) else [station['transline']]:
                transfer_station = "{} - {} 환승".format(station['name'], line_code)
                wait_time = calculate_wait_time(current_time)
                subway_graph[station['name']][transfer_station] = transfer_penalty + wait_time
                if transfer_station not in subway_graph:
                    subway_graph[transfer_station] = {}
                subway_graph[transfer_station][station['name']] = 0  # 환승 시 다시 돌아가는 경우는 없음

def dijkstra(graph, start, end):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    previous_nodes = {node: None for node in graph}
    priority_queue = []
    heapq.heappush(priority_queue, (0, start))

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_node == end:
            break

        for adjacent, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[adjacent]:
                distances[adjacent] = distance
                previous_nodes[adjacent] = current_node
                heapq.heappush(priority_queue, (distance, adjacent))

    path, current = [], end
    transfer_details = []
    while previous_nodes[current] is not None:
        path.insert(0, current)
        if "환승" in current:
            transfer_details.append(f"환승 at {current.split(' - ')[0]}")
        current = previous_nodes[current]
    if path:
        path.insert(0, current)
    return path, distances[end], transfer_details

path, total_time, transfers = dijkstra(subway_graph, '사당', '논현')
print("경로:", " -> ".join(path))
print("총 소요 시간:", total_time, "분")
if transfers:
    print("환승 정보:", ", ".join(transfers))
```
