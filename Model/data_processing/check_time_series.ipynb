```python
import os
import pandas as pd

folder_path = '/Users/vvoo/Capstone_TS/data/api/'
folders = sorted([f for f in os.listdir(folder_path) if f.startswith('2024')])
```

```python
folders
```

```python
df = pd.DataFrame()
for folder in folders:
    path = folder_path + folder
    try:
        tmp = pd.read_csv(f'{path}/combined_all.csv', index_col=0)
        df = pd.concat([df, tmp], ignore_index=False)
    except:
        pass

time_list = [str(i) for i in range(5,24)]
time_list.extend(['역명', '호선','상하행'])
df = df[time_list]
df
```

```python
# 역명 / 상행 / 하행
sejong = df[df['역명'] == '남부터미널']
sejong_up = sejong[sejong['상하행'] == '상행']
sejong_down = sejong[sejong['상하행'] == '하행']
```

```python
sejong_up
```

```python
import datetime

result_df = pd.DataFrame()
for idx in range(3, len(sejong_up)-1):
    tmp = pd.DataFrame(sejong_up.iloc[idx][:19])
    y, m, d = map(int, sejong_up.index[idx].split('-'))
    tmp.index = [datetime.datetime(y, m, d, int(hour)) for hour in tmp.index]
    tmp.rename(columns={sejong_up.index[idx]:'Duration'}, inplace=True)

    try:
        result_df = pd.concat([result_df, tmp], ignore_index=False, axis=0)
    except:
        result_df = tmp.copy()
    # display(tmp)
result_df
```

```python
result_df.fillna(method='ffill', inplace=True)
```

```python
import matplotlib.pyplot as plt

# 데이터가 이미 준비되어 있다고 가정
# 데이터를 시각화합니다.
plt.figure(figsize=(20, 10))  # 그래프 크기 조정
plt.plot(result_df.index, result_df['Duration'], linestyle='-', markersize=10)  # 마커 크기 조정
plt.title('Nambu Ter.', fontsize=16)
plt.xlabel('Date and Time', fontsize=14)
plt.ylabel('Duration', fontsize=14)
plt.grid(True)  # 그리드 추가
plt.xticks(rotation=45, fontsize=12)  # X축 레이블 회전 및 폰트 크기 조정
plt.yticks(fontsize=12)  # Y축 레이블 폰트 크기 조정
plt.tight_layout()
plt.show()
```

```python
result_df.to_csv('time_series.csv', index=True, encoding='utf-8-sig')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 가정된 데이터 구조를 DataFrame으로 생성
data = {
    'apiCalltime': [f"2024-04-27 {hour}:00" for hour in range(5, 24)] + [f"2024-04-28 {hour}:00" for hour in range(5, 24)],
    'Duration': [80, 92.166667, 90.3, 88.857143, 80.125, 90.4, 98.833333, 89.222222, 89.111111, 87.8, 71.25, 84.545455, 88.833333, 97.785714, 87.0, 78.333333, 53.2, 87.142857, 80.111111,
                 80, 92.166667, 90.3, 88.857143, 80.125, 90.4, 98.833333, 89.222222, 89.111111, 87.8, 71.25, 84.545455, 88.833333, 97.785714, 87.0, 78.333333, 53.2, 87.142857, 80.111111]
}

df = pd.DataFrame(data)
df['apiCalltime'] = pd.to_datetime(df['apiCalltime'])
df['Date'] = df['apiCalltime'].dt.date
df['Hour'] = df['apiCalltime'].dt.hour

# 데이터를 날짜와 시간으로 그룹화하여 평균을 계산
grouped = df.groupby(['Date', 'Hour'])['Duration'].mean().unstack()

# 그래프 그리기
grouped.plot(kind='bar', figsize=(15, 8))
plt.title('일별 및 시간대별 평균 소요 시간')
plt.xlabel('날짜')
plt.ylabel('평균 소요 시간(분)')
plt.xticks(rotation=0)
plt.legend(title='시간대', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# Set the date as the index
# df.set_index('apiCalltime', inplace=True)

# Fill NaN values, here I use forward fill as an example
df.fillna(method='ffill', inplace=True)

# Reshape the DataFrame for easier plotting (melt function)
df_melted = df.melt(id_vars=['역명', '호선', '상하행'], var_name='Hour', value_name='Duration')

# Create a plot
plt.figure(figsize=(15, 7))
for label, grp in df_melted.groupby(['역명', '호선', '상하행']):
    plt.plot(grp['Hour'], grp['Duration'], marker='o', label=f'{label}')

plt.title('Hourly Duration by Station and Direction')
plt.xlabel('Hour')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45)
plt.legend(title='Station Info', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
